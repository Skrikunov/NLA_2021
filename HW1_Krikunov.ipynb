# **1.1**
**- (1 pts) What are the constants $C_1$ and $C_2$ such that $C_1 \|x\|_{1} \leq \|x\|_{\infty} \leq C_2 \| x\|_{1}$ for any vector $x$**

Suppose that the vector $x$ is from $n$-dimentional vector space. Let's consider presented norms in details: $\|x\|_{1} \stackrel{\mathrm{def}}{=} \sum\limits_{i=1}^{n} |x_i|$; $ \|x\|_{\infty} \stackrel{\mathrm{def}}{=} \max_i | x_i| $

1) Let's find $C_2$. Suppose that the maximum element of vector $x$ is $x_k$, so $\|x\|_{\infty} = |x_k|$.

Let's rewrite the first norm as: $\|x\|_{1} = \sum\limits_{i=1,i\neq k}^{n} |x_i| + |x_k|$

There are 2 edge cases:
* $|x_i|$ are equal to each other.
* $|x_i| = 0, \forall i: i \neq k$

So, the following nonequality is true: $\|x\|_{1} = \sum\limits_{i=1,i\neq k}^{n} |x_i| + |x_k| \geq |x_k| = \|x\|_{\infty} \Rightarrow \|x\|_{\infty} \leq \|x\|_{1}, \frac{\|x\|_{\infty}}{ \|x\|_{1} }\leq 1, \forall x$, therefore $C_2 \geq 1$.

2) Let's find $C_1$.

$C_1 \|x\|_{1} \leq \|x\|_{\infty} \Rightarrow C_{1} \sum\limits_{i=1}^{n} |x_i| \leq |x_k| \Rightarrow C_{1} \leq \frac{|x_k|}{\sum\limits_{i=1}^{n}|x_i|}$

There are 2 edge cases:
* $|x_i|$ are equal to each other $\Rightarrow C_1 \leq \frac{x_k}{nx_k}=\frac{1}{n}$ 
* $|x_i| = 0, \forall i: i \neq k \Rightarrow C_1 \leq \frac{|x_k|}{|x_k|} = 1$ 

Therefore we can select $C_1 \leq \frac{1}{n}$ as the least option.

**Answer:** $C_1 \leq \frac{1}{n}, C_2 \geq 1$

**- (5 pts) Using the results from the previous subproblem, prove that $\| A \|_F \le \sqrt{\mathrm{rank}(A)} \| A \|_2$. _Hint:_ SVD will help you.**

Spectral norm definition: $ \|A \|_2 \stackrel{\mathrm{def}}{=} \sqrt {\lambda _{max}(A^{*}A)} \stackrel{\mathrm{def}}{=} \sigma_1(A) = \sigma_1$ - the largest singular value.

Let's consider SVD decomposition of the matrix $A$ (its frobenius norm): $\|A\|_F^2 \stackrel{\mathrm{1}}{=} \|U \Sigma V^*\|_F^2 \stackrel{\mathrm{2}}{=} \|\Sigma\|_F^2 \stackrel{\mathrm{3}}{=} \mathrm{trace}(\Sigma^*\Sigma) \stackrel{\mathrm{4}}{=} \sum\limits_{i=1}^{n} \sigma_i^2$
1. SVD decomposition: $A = U \Sigma V^*$, where $U$ and $V^*$ are unitary.
2. Unitary invariance: $ \| UAV\|_F = \| A \|_F.$
3. $\| A \|_F^2 = {\mathrm{trace}(A^*A)}$ - Frobenius norm property.
4. $\Sigma$ is diagonal matrix, therefore $\Sigma^*\Sigma$ is also diagonal with squared diag. elements.

There are two edge cases (important property: $\sigma_1 \geq \cdots \geq \sigma_r, \sigma_{r+1} \cdots \sigma_{n} = 0$, where $r=$ rank($A$),$n$ - an amount of singular values):
* $\sigma_i = 0, \forall i \geq 1 $ - single non-zero singular value. $\sum\limits_{i=1}^{n} \sigma_i^2 = \sigma_1$.
* $\sigma_i = \sigma_1, \forall i: i \leq r$ - all non-zero singular values are equal to each other. $\sum\limits_{i=1}^{n} \sigma_i^2 = r\sigma_1$

Obviously, the following non-equality is always true: $\sum\limits_{i=1}^{n} \sigma_i^2 \leq n\sigma_1$, that equal to: $\| A \|_F^2 \le {\mathrm{rank}(A)} \| A \|_2^2$ or $\| A \|_F \le \sqrt{\mathrm{rank}(A)} \| A \|_2$.

### Second step: neural network architecture

For simplicity and demonstration purposes of the neural network compression idea consider the architecture consisting of the only fully-connected layers and non-linear ReLU functions between them.
To demonstrate compression effect, consider the dimension of the inner layers equals to 1000.

Below you see implementation of such neural network in PyTorch.
More details about neural networks you will study in the *Deep learning* course in one of the upcoming term

````python
class ResidualBlock(nn.Module):
    def __init__(self, in_features, middle_features, activation=nn.ReLU()):
        super().__init__()
        self.layer1 = nn.Linear(in_features, middle_features)
        self.layer2 = nn.Linear(middle_features, in_features)
        self.activation = activation
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.activation(out)
        out = self.layer2(out)
        out = out + x
        out = self.activation(out)
        return out

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.res1 = ResidualBlock(1000,1000)
        self.res2 = ResidualBlock(1000,1000)
        self.fc2 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x):
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)
        x = self.res1(x)
        x = self.res2(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
````

#### Implement functions for training and testing after every sweep over all dataset entries

````python
def train(model, train_loader, optimizer, epoch):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.train()
    model.to(device)
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    device = "cuda" if torch.cuda.is_available() else "cpu"
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
````

````python
# this function is copy of test, but it doesnt print the result and it returns it. 

def get_accuracy(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    device = "cuda" if torch.cuda.is_available() else "cpu"
    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)
    
    return correct/len(test_loader.dataset)
````

````python
def SVD_layer(layer, compression_rank):
    r = compression_rank
    U, S, V = torch.linalg.svd(layer.weight.data.T) # layer SVD
    
    # u,s,v,bias as learnable parameters
    u = torch.nn.Parameter(U[:,:r], requires_grad = True)
    s = torch.nn.Parameter(S[:r], requires_grad = True)
    v = torch.nn.Parameter(V[:r,:], requires_grad = True)
    b = torch.nn.Parameter(layer.bias.data, requires_grad = True)
    return u, s, v, b   

class SVD_linear_layer(nn.Module):
    def __init__(self, base_layer, compression_rank):
        super().__init__()
        self.U, self.S, self.V, self.b = SVD_layer(base_layer, compression_rank)

    def forward(self, x):
        A = self.U @ torch.diag(self.S) @ self.V # construct low-rank weighting matrix
        x = x @ A + self.b # simulate NN layer
        return x

class SVD_ResidualBlock(nn.Module):
    def __init__(self, initial_block, compression_rank):
        super().__init__()
        self.activation = nn.ReLU()
        self.layer1 = SVD_linear_layer(initial_block.layer1, compression_rank)
        self.layer2 = SVD_linear_layer(initial_block.layer2, compression_rank)
        
    def forward(self, x):
        out = self.layer1(x)
        out = self.activation(out)
        out = self.layer2(out)
        out = out + x
        out = self.activation(out)
        return out   
    
class Compress_Net(nn.Module):
    def __init__(self,initial_net,compression_rank):
        super().__init__()
        self.fc1 = SVD_linear_layer(initial_net.fc1, compression_rank)
        self.res1 = SVD_ResidualBlock(initial_net.res1, compression_rank)
        self.res2 = SVD_ResidualBlock(initial_net.res2, compression_rank)
        self.fc2 = nn.Linear(1000, 10) # the last layer is regular NN linear layer
        self.ReLU = nn.ReLU()
        # copy params from original net to linear layer
        with torch.no_grad():
            self.fc2.weight.copy_(initial_net.fc2.weight.data)
            self.fc2.bias.copy_(initial_net.fc2.bias.data)

    def forward(self, x):
        x = self.fc1(x.view(-1,3*32*32))
        x = self.ReLU(x)
        x = self.res1(x)
        x = self.res2(x)
        x = self.fc2(x)
        return F.log_softmax(x, dim=1)
````

````python
import jax.numpy as jnp
import jax
import numpy as np
import random

class Game:
    def __init__(self, key):
        # key is a jax.random.PRNGKey
        self.key = key
        return
    def hermione_chooses_r_and_W(self):
        self.r = random.randint(1, 95)
        self.key, subkey = jax.random.split(self.key)
        self.W1 = jax.random.uniform(subkey, (self.r, 100), maxval=100., minval=0., dtype=jnp.float32)
        self.key, subkey = jax.random.split(self.key)
        self.W2 = jax.random.uniform(subkey, (100, self.r), maxval=100., minval=0., dtype=jnp.float32)
    def hermione_computes_function(self, x):
        return jnp.sum(jnp.square(jnp.sin(self.W2 @ jnp.cos(self.W1 @ x))))
    def harry_answers(self):
        # <your code here>
        # you shouldn't use self.r, self.W1, or self.W2
        # you can call `hermione_computes_function` multiple times
        r = 1 # for example

        gradient_F = jax.grad(self.hermione_computes_function, argnums=0)        
        # initialize gradient matrix
        gradient_matrix = jnp.zeros((100,100))
        # get 95 DIFFERENT examples (the least number. Otherwise we will have errors)
        for idx in range(95):
            # use different key for each vector
            x = jax.random.normal(jax.random.PRNGKey(idx),(100,)).astype(np.float32) 
            # calculate gradient
            grad = gradient_F(x)
            # update gradient matrix with new example
            gradient_matrix = gradient_matrix.at[:,idx].set(grad)
            
        U,S,V = jnp.linalg.svd(gradient_matrix/jnp.sqrt(100))
        # normalization for conveinience
        S /= S[0]
        # tolerance level for singular values to be taken into account
        # it has been estimated emperically
        tol = 1e-6
        # calculate all significant singular values
        r = jnp.sum(S > tol)
        return r
    
    def play(self, n_rounds, verbose=True):
        # n_rounds: a number of rounds of the game
        # verbose: print or not the result of each round
        n_right_answers = 0
        for _ in range(n_rounds):
            self.hermione_chooses_r_and_W()
            r = self.harry_answers()
            if abs(r - self.r) == 0:
                if verbose:
                    print("Good job! The true answer is {}, Harry's answer is {}!".format(self.r, r))
                n_right_answers += 1
            else:
                if verbose:
                    print("Harry's answers is {}, but the true answer is {} :(".format(r, self.r))
        if float(n_right_answers) / n_rounds > 0.95:
            print('Well done: {}/{} right answers!'.format(n_right_answers, n_rounds))
        else:
            print('Only {}/{} right answers :( Work a little more and you will succeed!'.format(
                n_right_answers, n_rounds))
````
